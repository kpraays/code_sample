```python
import csv
import datetime
import glob

save_path = "save_graphs\\attribute_files\\"

tweet_attributes = save_path+"tweet_attributes.tsv"
user_attributes = save_path+"user_attributes.tsv"
tweets_state = save_path+"tweet_statewise_unique_tweet_id.tsv"
tweets_keywords = save_path+"tweet_keywords.tsv"
```

```python
def look_at_all_files(func):
    def inner(*args, **kwargs):
        printl(f"looking at .tsv files in directory {args[0]} for [{args[1]}]")
        files = glob.glob(args[0]+'[hydrated-tweets]*.tsv')
        for file in files:
            kwargs["file_name"]=file
            func(*args, **kwargs)
    return inner

def printl(text):
    with open(save_path+"attribute-files-logs.txt","a") as log_file:
        log_file.write(f"{datetime.datetime.now().strftime('%c')}   {text}")
        log_file.write("\n")
        
def total_records(file_name):
    """
    returns the total number of tweets in the file
    we are having a single tweet per row - each having a unique tweet id
    """
    with open(file_name,"r") as node_list:
        records=0
        for rows in node_list:
            records+=1
        return records
```

```python
@look_at_all_files
def tweet_attribute_file(directory, purpose, **kwargs):
    file_name = kwargs["file_name"]
    printl(f"total records in {file_name} are {total_records(file_name)}")
    
    with open(file_name,"r") as tweet_list:
        with open(tweet_attributes, "a") as tweet_attribute:
            f = csv.writer(tweet_attribute,delimiter="\t")
            for rows in tweet_list:
                
                cols = rows.split("\t")
                if cols[0] == "tweet_id":
                    headers = ["tweet_id", "tweet_content", "in_reply_to_status_id_str", "created_at", "retweet_count", "favorite_count","label"]                
                    f.writerow(headers)
                    continue
                f.writerow([cols[0],cols[1],cols[3],cols[5],cols[6],cols[7]])
```

```python
@look_at_all_files
def user_attribute_file(directory, purpose, **kwargs):
    file_name = kwargs["file_name"]
    printl(f"total records in {file_name} are {total_records(file_name)}")
    
    with open(file_name,"r") as user_list:
        with open(user_attributes, "a") as user_attribute:
            f = csv.writer(user_attribute,delimiter="\t")
            for rows in user_list:
                
                cols = rows.split("\t")
                if cols[0] == "tweet_id":
                    headers = ["user_id", "in_reply_to_user_id_str", "user_followers_count", "user_friends_count", "user_listed_count","label"]   
                    f.writerow(headers)
                    continue
                f.writerow([cols[2],cols[4],cols[8],cols[9],int(cols[10])])
```
